```python
sample=dt.iloc[:10,:]
```

```python
plt.subplots(nrows=1, ncols=2, figsize=(25, 6))

plt.subplot(1, 2, 1)
plt.scatter(sample.Country, sample['2022 Population'])
plt.title('Population of Countries in 2022', pad=20)
plt.xlabel('Continents', labelpad=10, fontdict={'fontweight': 'bold'})
plt.ylabel('population percentage')
plt.yticks([200, 400, 600, 800, 1000, 1200],
            ['0.10', '0.20', '0.30', '0.40', '0.50', '0.60'])
plt.xticks(rotation=45)


plt.subplot(1, 2, 2)
plt.scatter(sample.Country, sample['1980 Population'])
plt.title('Population of Countries in 1980', pad=20)
plt.xlabel('Continents', labelpad=10, fontdict={'fontweight': 'bold'})
plt.ylabel('population in year')
plt.yticks([200, 800, 1400, 2000, 2600, 3200, 3800],
            ['0.2B', '0.4B', '0.6B', '0.8B', '1.0B', '1.2B', '1.4B'])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x="Country", y="2022 Population", data=sample)
plt.title('Population Projections For 2020')
plt.ylabel('population in billions')

plt.yticks([200, 400, 600, 800, 1000, 1200, 1400, 1600],
            ['0.2B', '0.4B', '0.6B', '0.8B', '1.0B', '1.2B', '1.4B', '1.6B'])
plt.show()
```
